```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset/Iris.csv")
```

```python
SepalLengthCm_frequency = df.SepalLengthCm.value_counts().to_frame()
SepalWidthCm_frequency = df.SepalWidthCm.value_counts().to_frame()
PetalLengthCm_frequency = df.PetalLengthCm.value_counts().to_frame()
PetalWidthCm_frequency = df.PetalWidthCm.value_counts().to_frame()
Species_frequency = df.Species.value_counts().to_frame()
```

```python
SepalLengthCm_frequency.reset_index(inplace=True)
SepalWidthCm_frequency.reset_index(inplace=True)
PetalLengthCm_frequency.reset_index(inplace=True)
PetalWidthCm_frequency.reset_index(inplace=True)
Species_frequency.reset_index(inplace=True)
```

```python
SepalLengthCm_frequency.rename(columns = {'index':'SepalLengthCm', 'SepalLengthCm': 'total'}, inplace=True)
SepalWidthCm_frequency.rename(columns = {'index':'SepalWidthCm', 'SepalWidthCm': 'total'}, inplace=True)
PetalLengthCm_frequency.rename(columns = {'index':'PetalLengthCm', 'PetalLengthCm': 'total'}, inplace=True)
PetalWidthCm_frequency.rename(columns = {'index':'PetalWidthCm', 'PetalWidthCm': 'total'}, inplace=True)
Species_frequency.rename(columns = {'index':'Species', 'Species': 'total'}, inplace=True)
```

```python
SepalLengthCm_frequency.total.astype('int64')
SepalWidthCm_frequency.total.astype('int64')
PetalLengthCm_frequency.total.astype('int64')
PetalWidthCm_frequency.total.astype('int64')
Species_frequency.total.astype('int64')
```

```python
plt.figure(figsize=(30, 10))
plt.bar(Species_frequency['Species'], Species_frequency['total'], align='center')
plt.xticks(Species_frequency['Species'])
plt.show()

plt.figure(figsize=(30, 10))
plt.bar(SepalLengthCm_frequency['SepalLengthCm'], SepalLengthCm_frequency['total'], width=0.1, align='center')
plt.xticks(SepalLengthCm_frequency['SepalLengthCm'])
plt.show()

plt.figure(figsize=(30, 10))
plt.bar(SepalWidthCm_frequency['SepalWidthCm'], SepalWidthCm_frequency['total'], width=0.1, align='center')
plt.xticks(SepalWidthCm_frequency['SepalWidthCm'])
plt.show()

plt.figure(figsize=(30, 10))
plt.bar(PetalLengthCm_frequency['PetalLengthCm'], PetalLengthCm_frequency['total'], width=0.1, align='center')
plt.xticks(PetalLengthCm_frequency['PetalLengthCm'])
plt.show()

plt.figure(figsize=(30, 10))
plt.bar(PetalWidthCm_frequency['PetalWidthCm'], PetalWidthCm_frequency['total'], width=0.1, align='center')
plt.xticks(PetalWidthCm_frequency['PetalWidthCm'])
plt.show()
```

```python
df_analysis = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
df_estatistica = df_analysis.describe()
```

```python
index_list = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
new_dataframe = pd.DataFrame(columns=['mode', 'mean', 'median', 'kurtosis', 'skew', 'var'], index=index_list)
```

```python
mode = df_analysis.mode()
mean = df_analysis.mean()
median = df_analysis.median()
kurtosis = df_analysis.kurtosis()
skew = df_analysis.skew()
var = df_analysis.var()
```

```python
new_dataframe['mode'] = mode.values.tolist()[0]
new_dataframe['mean'] = mean.tolist()
new_dataframe['median'] = median.tolist()
new_dataframe['kurtosis'] = kurtosis.tolist()
new_dataframe['skew'] = skew.tolist()
new_dataframe['variance'] = var.tolist()
new_dataframe['std'] = df_estatistica.loc[['std']].values.tolist()[0]
new_dataframe['25%'] = df_estatistica.loc[['25%']].values.tolist()[0]
new_dataframe['50%'] = df_estatistica.loc[['50%']].values.tolist()[0]
new_dataframe['75%'] = df_estatistica.loc[['75%']].values.tolist()[0]
new_dataframe
```
